Modeling

```python
#Normalize the amount column
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
credit_df['scaled_amount']=scaler.fit_transform(credit_df['Amount'].values.reshape(-1,1))
credit_df['scaled_time'] = scaler.fit_transform(credit_df['Time'].values.reshape(-1,1))
credit_df.drop(['Time','Amount'], axis=1, inplace=True)
credit_df.head()
```

```python
#assign dependent and independent features
X=credit_df.drop('Class',1)
y=credit_df['Class']
```

```python
#print the shape of X and y
print(X.shape)
print(y.shape)
```

```python
#Splitting of data set into test and train
from sklearn.model_selection import  train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=1)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
#roc curve 
def compute_roc(y_test,y_pred,plot=False):
  fpr, tpr, _ = roc_curve(y_test, y_pred)
  auc_score = auc(fpr, tpr)
  if plot:
    plt.figure(figsize=(7,6))
    plt.plot(fpr,tpr,color='red',label='ROC(AUC=%0.3f)'% auc_score)
    plt.legend(loc='lower right')
    plt.title('ROC Curve')
    plt.xlabel("FPR")
    plt.ylabel('TPR')
    print(f'\n AUC Score:',auc_score)
    return
```

```python
from sklearn.metrics import confusion_matrix, recall_score, f1_score, classification_report,precision_score,accuracy_score,roc_curve,auc
def model_performance(y_test,y_pred):
  print("\n Confusion Matrix:\n{}".format(confusion_matrix(y_test, y_pred)))
  print("\n Recall Score:{}".format(recall_score(y_test, y_pred)))
  print("\n Precision Score:{}".format(precision_score(y_test, y_pred)))
  print("\n Accuracy Score:{}".format(accuracy_score(y_test,y_pred)))
  print("\n F1 Score: {}".format(f1_score(y_test,y_pred)))
  compute_roc(y_test,y_pred,plot=True)
  print('\n')
  return
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
logreg=LogisticRegression()
logreg.fit(X_train,y_train)
y_pred=logreg.predict(X_test)
print('Logistic Regression:\n')
model_performance(y_test,y_pred)

svc=SVC()
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
print('Support Vector Classifier:\n')
model_performance(y_test,y_pred)

rf=RandomForestClassifier()
rf.fit(X_train,y_train)
y_pred=rf.predict(X_test)
print('Random Forest:\n')
model_performance(y_test,y_pred)
```
